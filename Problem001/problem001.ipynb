```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
text_array = np.loadtxt('avg.txt')
```

```python
x = text_array[:, 0]
y = text_array[:, 1]
```

```python
fig, ax = plt.subplots(figsize = (9, 7))
ax.scatter(x, y, s=80)
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("Rozmiar obrazka [piksel x piksel]")
ax.set_ylabel("Åšredni czas generowania obrazka [s]")

plt.savefig('plot.png')
```
